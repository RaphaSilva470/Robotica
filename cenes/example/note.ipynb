```python
from coppeliasim_zmqremoteapi_client import RemoteAPIClient
```

```python
client = RemoteAPIClient()
sim = client.require('sim')

sim.setStepping(True)

sim.startSimulation()
while (t := sim.getSimulationTime()) < 10:
    print(f'Simulation time: {t:.2f} [s]')
    sim.step()
sim.stopSimulation()
```
